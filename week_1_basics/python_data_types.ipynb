## Variable Assignments and Types

```python
# this is a comment
x = 3 # this is a variable (x) assigned to the integer value 3
```

```python
# type() is a function that can tell us the data type of a variable
type(x)
```

```python
# what about decimals?
y = 3.5
type(y)
```

```python
z = x + y
z
```

```python
type(z)
```

Sometimes, Python isn't as smart as you think it is (or should be)

```python
my_text+z
```

We can do a unit conversion, though

```python
my_text + str(z)
```

We can also use Python's *formatting* commands to add a non-text-value to a string (see: https://docs.python.org/3/tutorial/inputoutput.html https://docs.python.org/3/reference/lexical_analysis.html#f-strings https://docs.python.org/3/library/string.html#formatspec)

```python
f"{my_text} {z:.2f}"
```

```python
f"{my_text} {z:010.2f}"
```

```python
f"{my_text} {z:5.2f}"
```
