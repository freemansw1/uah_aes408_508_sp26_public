```python
huntsville_location = (34.72806, -86.64733)
armor_location = (34.646166098419094, -86.77144334127473)
m = folium.Map(location=huntsville_location, zoom_start=12, tiles="cartodb positron")
folium.Marker(
    location=armor_location,
    tooltip="Click me!",
    popup="Mt. Hood Meadows",
    icon=folium.Icon(icon="cloud"),
).add_to(m)

m
```
